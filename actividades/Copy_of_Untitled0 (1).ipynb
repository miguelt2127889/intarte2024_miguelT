```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

# Dividir datos en conjuntos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=0)

# Entrenar el modelo de regresión lineal
regression_model = LinearRegression()
regression_model.fit(train_data, train_labels)

# Realizar predicciones
predictions = regression_model.predict(test_data)

# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predictions)

# Graficar el modelo de regresión lineal
plt.scatter(test_data, test_labels, color='blue')
plt.plot(test_data, predictions, color='red')
plt.title('Modelo de Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.show()

# Mostrar el error cuadrático medio
print("Error cuadrático medio:", mse)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2

train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=0)

regression_model = LinearRegression()
regression_model.fit(train_data, train_labels)

predictions = regression_model.predict(test_data)

mse = mean_squared_error(test_labels, predictions)

plt.scatter(test_data, test_labels, color='blue')
plt.plot(test_data, predictions, color='red')
plt.title('Modelo de Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.show()

print("Error cuadrático medio:", mse)
```
